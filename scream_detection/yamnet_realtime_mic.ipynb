```python
import pyaudio
import librosa
import numpy as np
import matplotlib.pyplot as plt
import keras

import params as yamnet_params
import yamnet as yamnet_model
```

```python
params = yamnet_params.Params()
yamnet = yamnet_model.yamnet_frames_model(params)
yamnet.load_weights('yamnet.h5')
yamnet_classes = yamnet_model.class_names('yamnet_class_map.csv')
```

```python
def listening():
    frame_len = int(params.sample_rate * 1) # 1sec
    cnt = 0
    p = pyaudio.PyAudio()
    stream = p.open(format=pyaudio.paInt16,
                channels=1,
                rate=int(params.sample_rate),
                input=True,
                frames_per_buffer=frame_len)
    # plt.ion()
    while True:
        try:
            # data read
            data = stream.read(frame_len, exception_on_overflow=False)

            # byte --> float
            frame_data = librosa.util.buf_to_float(data, n_bytes=2, dtype=np.int16)

            # model prediction
            scores, embeddings, spectrogram = yamnet(frame_data)
            scores = scores.numpy()
            #spectrogram = spectrogram.numpy()

            mean_scores = np.mean(scores, axis=0)

            top5_i = np.argsort(mean_scores)[::-1][:3]

        #     print(cnt)
            cnt+=1

#             print('Current event:\n' +
#                   '\n'.join('  {:12s}: {:.3f}'.format(yamnet_classes[i], mean_scores[i])
#                             for i in top5_i))

            classes = []
            for i in top5_i:
                classes.append(yamnet_classes[i])
                print(cnt, classes)
#                 if yamnet_classes[i] in ["Yell", "Screaming", "Crying, sobbing", 'Wail, moan', 'Groan', 'Burping, eructation']:
#                     print(cnt, yamnet_classes[i])
#                     break
            else:
                continue
                #print(cnt," Irrelevant")

        except (KeyboardInterrupt, EOFError, SystemExit):
            # Press ctrl-c or ctrl-d on the keyboard to exit
            stream.stop_stream()
            stream.close()
            p.terminate()
            break

#Start speaking into mic, if scream or pain-related sounds detected, it will say Relevant.
```

```python
# Find the name of the class with the top score when mean-aggregated across frames.
def class_names_from_csv(class_map_csv_text):
    """Returns list of class names corresponding to score vector."""
    class_map_csv = io.StringIO(class_map_csv_text)
    class_names = [display_name for (class_index, mid, display_name) in csv.reader(class_map_csv)]
    class_names = class_names[1:]  # Skip CSV header
    return class_names
```

```python
def listening():
    frame_len = int(params.sample_rate * 3) # 1sec
    cnt = 0
    p = pyaudio.PyAudio()
    stream = p.open(format=pyaudio.paInt16,
                channels=1,
                rate=int(params.sample_rate),
                input=True,
                frames_per_buffer=frame_len)
    interpreter = tf.lite.Interpreter('yamnet.tflite')

    input_details = interpreter.get_input_details()
    waveform_input_index = input_details[0]['index']
    output_details = interpreter.get_output_details()
    scores_output_index = output_details[0]['index']
    embeddings_output_index = output_details[1]['index']
    spectrogram_output_index = output_details[2]['index']
    
    class_names = class_names_from_csv(open('yamnet_class_map.csv').read())

    while True:
        try:
            # data read
            data = stream.read(frame_len, exception_on_overflow=False)

            # byte --> float
            waveform = librosa.util.buf_to_float(data, n_bytes=2, dtype=np.int16)
            
            if len(waveform.shape) > 1:
                    waveform = np.mean(waveform, axis=1)
                    
            waveform = np.array(waveform, dtype=np.float32)

            interpreter.resize_tensor_input(waveform_input_index, [len(waveform)], strict=True)
            interpreter.allocate_tensors()
            interpreter.set_tensor(waveform_input_index, waveform)
            interpreter.invoke()
            scores, embeddings, spectrogram = (
                interpreter.get_tensor(scores_output_index),
                interpreter.get_tensor(embeddings_output_index),
                interpreter.get_tensor(spectrogram_output_index))
            #print(scores.shape, embeddings.shape, spectrogram.shape)  # (N, 521) (N, 1024) (M, 64)

            #print(class_names[scores.mean(axis=0).argmax()])
            mean_scores = np.mean(scores, axis=0)

            top5_i = np.argsort(mean_scores)[::-1][:3]

            cnt+=1
            classes = []
            for i in top5_i:
                classes.append(class_names[i])
#                 if yamnet_classes[i] in ["Yell", "Screaming", "Crying, sobbing", 'Wail, moan', 'Groan', 'Burping, eructation']:
#                     print(cnt, yamnet_classes[i])
#                     break
#             else:
#                 continue
                #print(cnt," Irrelevant")
            print(cnt, classes)

        except (KeyboardInterrupt, EOFError, SystemExit):
            # Press ctrl-c or ctrl-d on the keyboard to exit
            stream.stop_stream()
            stream.close()
            p.terminate()
            break

#Start speaking into mic, if scream or pain-related sounds detected, it will say Relevant.
```
